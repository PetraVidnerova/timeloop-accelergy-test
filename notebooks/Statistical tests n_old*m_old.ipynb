```python
import numpy as np 
import pandas as pd
```

```python
PREFIX="simba_"
#PREFIX="eyeriss_"
```

```python
df = pd.read_csv(f"data/{PREFIX}energy_inputs.csv", sep=";", header=None)
df.columns = ["n_in", "timeloop_energy", "timeloop_comp", "timeloop_data", "jirka_comp", "jirka_data"]
df
```

```python
df = df[["n_in", "timeloop_data"]]
df["n_in_sq"] = df["n_in"]*df["n_in"]
```

```python
df
```

# Statistical test using smf.osl

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula='timeloop_data ~ n_in + n_in_sq', data=df)
results = model.fit()
```

```python
results
```

```python
results.summary()
```

```python
print(results.pvalues)
```

```python
results.pvalues["n_in_sq"]
```

```python
results.pvalues[results.pvalues > .05]
```

```python
f"{results.pvalues['n_in_sq']:.3f}"
```

# Figure

```python
from scipy import stats
m, c, r, p, se = stats.linregress(df["n_in_sq"], df["timeloop_data"])

print(m, c)

df["sq"] = m*df["n_in_sq"]+c
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,4))

dfx = df[["n_in", "timeloop_data", "sq"]]

sns.barplot(data=dfx, x="n_in", y="timeloop_data", color="gray")
sns.lineplot(data=dfx, x=dfx.index, y="sq", ls="--", color="black")

ax.set(xlabel="$n_{in}$")
ax.set(ylabel="estimated $E_{data}$")
#ax.set(ylim=(0,6e09))
ax.set(xlim=(0,50))
#ax.set_xticks(list(range(56,448,1)));
ax.set_xticks(range(0, 50, 5))

fig.tight_layout()
fig.savefig(f"{PREFIX}sq_trend_n.png", bbox_inches="tight")
```
